# Installation of Python and Pygame

We have seen in the Introduction section that the programming language `Python` and the graphics library `Pygame` are very powerful tools for designing psychological experiments. 
In this section we will outline the installation steps for both `Python` and `Pygame` regarding teh following operating systems:

- Windows
- Mac
- Linux

Please follow the instructions under the subsection describing the installation for your operating system.

## Windows
If you are using a Windows operating system follow these steps:

- Go to the [Python Website](https://www.python.org/downloads/windows/) and download a Python distribution for your system.

- Run the Installer

- Enable "Add Python 3.6 to PATH" and click "Install Now"

![alt text](WindowsInstallStart.png "Windows Installer")

- Thats it. Now the setup is finished and you should see the following:

![alt text](WindowsInstallEnd.png "Finish Installler")

- You can check if the installation was succesfull by opening the command line. The command line can be started by searching for "cmd" in the start menu. Once the command line is started type `python` and press `Return`.

You should see the following:

![alt text](WindowsPythonTerminal.png "Check Python Cmd")


- This means that the python installation was succesfull as you have started the python interpreter. To exit the interpreter type `quit()` and press `Return`.

- Next we will install Pygame. In the Terminal type the following and press `Return`:

```bash
python -m pip install pygame --user
```
- To check whether the installatio worked type the following and press `Return`:

```bash
python -m pygame.examples.aliens
```

- If the program starts then pygame is installed sucessfully and you should see the aliens game running.

## Mac OS
If you are using a Mac OS follow these steps:

- Open your terminal type the following and press `Return`:
```bash
xcode-select --install
```
- Install ruby by typing:
```bash
ruby -e "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/master/install)"
```
- To confirm the installation run:
```bash
brew doctor
```
This should print `Your system is ready to brew.`.<br><br>
- Now you can install python 3 via the command: 
```bash
brew install python3
```
- Check if it was succesfull via:
```bash
python3 --version 
```
This should print your python 3 version.<br><br>
- To install `Pygame` run:
```bash
pip install pygame
```
- To check whether the installation worked type the following and press `Return`:

```bash
python -m pygame.examples.aliens
```

- If the program starts then pygame is installed sucessfully and you should see the aliens game running.

Note that this is one of many ways how to install `Python` and `Pygame`. If this does not work for you please search on the internet for resources and guides on how to make the installation work.

## Linux

If you are using a Linux distribution follow these steps:

- Open your terminal, enter the follwing commands and press `Return` (this updates packages first):
```bash
sudo apt update
sudo apt install software-properties-common
```
- Run the follwing command to add daeadsnake PPA to sources list and press `Return` when prompted:
```bash
sudo add-apt-repository ppa:deadsnakes/ppa
```
- Now you can install pyhton from the repo:
```bash
sudo apt install python3.7
```
- Check if it was succesfull via:
```bash
python3 --version
```
This should print your python 3 version.<br><br>
- To install `Pygame` run:
